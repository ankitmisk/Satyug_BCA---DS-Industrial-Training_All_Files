# 1. Strip Plot
A strip plot is used to display the distribution of data points along a categorical axis.

````python


# Create a strip plot
# plt.figure(figsize=(10, 6))

sns.stripplot(data = tips, x = 'day',y = 'total_bill', hue = 'sex')

plt.title('Strip Plot of Total Bill by Day')
plt.show()
````

# 2. Distribution Plot (distplot)
Distribution plots are used to visualize the distribution of a dataset.

````python
# Create a distribution plot

sns.histplot(tips['total_bill'])

# plt.title('Distribution Plot of Total Bill')
plt.show()
````

````python
# Create a distribution plot

sns.histplot(tips['tip'])

# plt.title('Distribution Plot of Total Bill')
plt.show()
````

# 3. Joint Plot
Joint plots are used to visualize the relationship between two variables along with their marginal distributions.

````python
# Create a joint plot

# { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }


sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
# plt.suptitle('Joint Plot of Total Bill and Tip', y=1.02)
plt.show()
````

````python
# Create a joint plot

# { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }


sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
# plt.suptitle('Joint Plot of Total Bill and Tip', y=1.02)
plt.show()
````

# 4. Violin Plot
A violin plot combines aspects of a box plot and a KDE plot to provide a richer display of the distribution.

````python
# Create a violin plot
plt.figure(figsize=(10, 6))

sns.violinplot(x='day', y='total_bill', data=tips,hue = 'day')
plt.title('Violin Plot of Total Bill by Day')

plt.show()
````

````python
# Create a violin plot
plt.figure(figsize=(10, 6))

sns.boxplot(x='day', y='total_bill', data=tips,hue = 'day')
plt.title('Violin Plot of Total Bill by Day')

plt.show()
````

# 5. Swarm Plot
A swarm plot is similar to a strip plot but the points are adjusted to avoid overlapping.

````python
# Create a swarm plot
plt.figure(figsize=(10, 6))

sns.swarmplot(x='day', y='total_bill', data=tips)
plt.title('Swarm Plot of Total Bill by Day')

plt.show()
````

# 6. Pair Plot
Pair plots are used to visualize pairwise relationships in a dataset.

````python
# Create a pair plot
sns.pairplot(tips)
plt.suptitle('Pair Plot of Tips Dataset', y=1.02)
plt.show()
````

# 7. Count Plot
A count plot is used to show the counts of observations in each categorical bin.

````python
# Create a count plot
plt.figure(figsize=(10, 6))

sns.countplot(x='day', data=tips)
plt.title('Count Plot of Days')

plt.show()
````

# 8. Heatmap
A heatmap is used to display the magnitude of values as colors in a matrix.

````python
# Create a heatmap
corr = tips.corr(numeric_only=True)
plt.figure(figsize=(10, 6))

sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

plt.title('Heatmap of Tips Dataset')
plt.show()
````

# 9. Box Plot
A box plot displays the distribution of data based on a five-number summary: minimum, first quartile, median, third quartile, and maximum.

````python
# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.show()
````

# 10. Bar Plot
A bar plot is used to show the mean or other aggregation of a numeric variable across the levels of a categorical variable.

````python
# Create a bar plot
plt.figure(figsize=(10, 6))

sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot of Total Bill by Day')

plt.show()
````

# 11. Scatter Plot
A scatter plot is used to visualize the relationship between two numeric variables.

````python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill and Tip')
plt.show()
````

# 12. lmplot
An lmplot is used to plot data and regression models across different subsets of the data.

````python
# Create a lmplot
plt.figure(figsize=(10, 6))
sns.lmplot(data = tips, x = 'total_bill', y = 'tip')
plt.title('Linear Model of Total Bill vs tip')
plt.show()
````
